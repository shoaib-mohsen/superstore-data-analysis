# SUMMARY

Sales is considered one of the most crucial determinants of superstores' performance and profitability, as they can reveal high-value insights about businesses' operations, helping identify deviations and correcting them to support data driven decisions and reach predefined goals.

This report presents a sales data analysis for a superstore operating in the United States, selling Furniture, Office Supplies and technology products to across the country. The dataset contains information on sales, ordering and shipping Dates, categories and sub categories, geographic location, shipping modes and other relevant id/names for orderes made between 2015 and 2018. 

analysis shows significant seasonal drops in sales each january, February. However, january .that comes alongside an overall increasing trend over time. Sales breakdown by category shows relatively similar sales volumes across categories. However, the top sub-categories by sales are phones, chairs, storage, tables and binders, and orders are more often made from NYC, LA and seattle.

deeper analysis of shipping duration by shipping mode reveals valuable insights as some same-day orders taking up to 1 day in some states like Arizona, California, Colorado, North Calorina, Texas and Washington.Contrariwise, First class orders taking less than one day in states like New Mexico, Lowa and New Hampshire. while other orders taking up to 4 days in Ohio. This pattern appears in second and standard classes with shipping durations reaching up to five to seven days in some states, while remaining shorter in others.

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from datetime import date, datetime
data = pd.read_csv('superstore_final_dataset.csv', encoding='latin1' , parse_dates = ['Order_Date' , 'Ship_Date' ])
data['Order_Date'] = pd.to_datetime(data['Order_Date'], dayfirst=True)
data['Ship_Date'] = pd.to_datetime(data['Ship_Date'], dayfirst=True)
```

```python
print(data.shape)
data['Order_Date' ].describe()
```

```python
data.describe()
```

```python
plt.figure(figsize=(15, 5))
monthly_data = data['Sales'].groupby(data['Order_Date'].dt.to_period('M')).sum()
monthly_data.index = monthly_data.index.astype(str)
plt.xticks(rotation=45 , ha='right')
plt.plot(monthly_data.index,monthly_data.values )
```

```python
data_by_category = data.groupby('Category')['Sales'].sum()
plt.pie(data_by_category.values ,labels = data_by_category.index ,  colors = ["#00B4D8", "#0077B6", "#03045E", "#14213D", "#2A9D8F", "#4EA8DE", "#90E0EF", "#A8DADC", "#457B9D", "#1D3557", "#E2EAFC"])
```

```python
data_by_city = pd.DataFrame(data.groupby('City')['Sales'].sum())
data_by_city = data_by_city.sort_values(by="Sales", ascending=False).reset_index()
top_n = data_by_city.iloc[:3].copy()
others_val = data_by_city.iloc[3:]["Sales"].sum()
others_df = pd.DataFrame([{"City": "Others", "Sales": others_val}])
final_df = pd.concat([top_n, others_df])
plt.figure(figsize=(6, 6))
plt.pie(
    final_df['Sales'],
    labels=final_df['City'],
    colors= ["#00B4D8", "#0077B6", "#14213D", "#E2EAFC", "#2A9D8F", "#4EA8DE", "#90E0EF", "#A8DADC", "#457B9D", "#1D3557"]
)
```

```python
data_by_subCategory = pd.DataFrame(data.groupby('Sub_Category')['Sales'].sum())
data_by_subCategory = data_by_subCategory.sort_values(by="Sales", ascending=False).reset_index()
top_n = data_by_subCategory.iloc[:5].copy()
others_val = data_by_subCategory.iloc[5:]["Sales"].sum()
others_df = pd.DataFrame([{"Sub_Category": "Others", "Sales": others_val}])
final_df = pd.concat([top_n, others_df])
plt.figure(figsize=(6, 6))
plt.pie(
    final_df['Sales'],
    labels=final_df['Sub_Category'],
    colors= ["#457B9D","#00B4D8", "#14213D", "#2A9D8F", "#4EA8DE","#E2EAFC", "#90E0EF", "#A8DADC", "#1D3557"]
)
```

```python
data['shipping_delay'] = data['Ship_Date'] - data['Order_Date']
data['shipping_delay'].describe()
```

```python
data['shipping_delay'].dt.days.astype('int64').plot( kind = 'hist' , color = 'blue' , bins = 8)
```

```python
data['shipping_delay'].value_counts()
```

```python
shippingDelay_shipMode_data = data[['shipping_delay' , 'Ship_Mode']].groupby('Ship_Mode').mean()
shippingDelay_shipMode_data.sort_values(by = 'shipping_delay' )
```

```python
data['shipping_delay_days'] = data['shipping_delay'].dt.total_seconds() / (24 * 3600)
geoDelay_data = data[['shipping_delay_days' , 'Ship_Mode','State']].groupby(['State', 'Ship_Mode']).max().reset_index()
pivot = geoDelay_data.pivot(
    index='State',
    columns='Ship_Mode',
    values='shipping_delay_days'
)

pivot = pivot[
    ['Same Day', 'First Class', 'Second Class', 'Standard Class']
]

plt.figure(figsize=(12,10))

sns.heatmap(
    pivot,
    cmap='YlOrRd',
    annot=True,
    fmt='.1f'
)

plt.title('Maximum Shipping Delay by State and Ship Mode')
plt.show()
```

```python
daily_data = data[[ 'Sales' , 'Order_Date']].groupby('Order_Date').sum().reset_index()
daily_data['Order_Month'] = daily_data['Order_Date'].dt.month
data_by_month = daily_data[[ 'Sales' , 'Order_Month']].groupby('Order_Month').mean().reset_index()
data_by_month.sort_values(by = 'Sales')
```
